```python
import sklearn, json
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
expert_dct = json.load(open('expert_all.json'))
```

```python
brk, cap, her, aei = [],[],[],[]
press = ['drumbeats', 'press']
testimony = ['testimony', 'testimonies']
events = ['events']


def extract_type(href):
    cut = ''
    
    if '?post_type=' in href: return
    if 'brookings.edu' in href: 
        cut = href.split('/')[3]; brk.append(cut)
    
    if 'heritage.org' in href:
        try:
            cut = href.split('/')[5]
            if len(cut.split('-'))>1: 
                cut = href.split('/')[3]
            if len(cut) == 0:
                cut = href.split('/')[4]
            her.append(cut); 
        except: return
    
    if 'americanprogress.org'in href: 
        cut = href.split('/')[3]; cap.append(cut);
        
    if 'aei.org'in href: 
        cut = href.split('/')[3]; aei.append(cut); 
    
    if cut in press: return 'press'
    if cut in testimony: return 'testimony'
    if cut in events: return
    if len(cut)!=0: return 'opinions'
    
    else: 
        print(href)
        #try: return(href[7:].strip('/|.|w').split('/')[0].split('.')[-2])
        #except IndexError: return(href[7:].strip('/|.|w').split('/')[0])
        return 'external'
    
def match(x, tp):
    try: return stat[x][tp]; 
    except KeyError: return 0
```

```python
stat = dict()
for exp in expert_dct:
    art = set([i for i in expert_dct[exp]['articles']])
    stat[exp] = Counter([extract_type(i) for i in art])
```

```python
data = pd.read_csv('expert_bio_complete.csv')
data['num_opinions'] = data['expert_id'].apply(lambda x: match(x, 'opinions'))
data['num_testimony'] = data['expert_id'].apply(lambda x: match(x, 'testimony'))
data['num_press'] = data['expert_id'].apply(lambda x: match(x, 'press'))
data['num_external'] = data['expert_id'].apply(lambda x: match(x, 'external'))
data.to_csv('expert_bio_complete.csv', index=False)
```
